**DATA SPLITTING**

```python
from sklearn.model_selection import train_test_split

train_data, test_data = train_test_split(df, test_size=0.3, random_state=42)

df['source'].value_counts()
```

```python
class Dataset(Dataset):
    def __init__(self, dataframe):
        self.texts = dataframe['text'].values.tolist()
        self.labels = dataframe['source'].values.tolist()

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):
        X = self.texts[idx]
        y = 1 if self.labels[idx] == 'ai' else 0  # Convert 'ai'/'human' to binary labels
        X = X.lower()
        tokens = X.split()  # Use basic whitespace tokenization
        X = " ".join(tokens).strip()
        return X, y
```

```python
def collate_fn(batch):
    x, y = [], []
    for text, label in batch:
        x.append(text)
        y.append(label)
    padded_text = tokenizer(x, padding=True, truncation=True, return_tensors='pt')

    return padded_text, torch.FloatTensor(y)
```

```python
class transformer(nn.Module):
    def __init__(self, base_model):
        super(transformer, self).__init__()

        self.bert = base_model
        self.fc = nn.Linear(768, 1)
        self.sigmoid = nn.Sigmoid()

    def forward(self, input_ids, attention_mask):
        bert_out = self.bert(input_ids=input_ids,
                             attention_mask=attention_mask)[0][:, 0]
        x = self.fc(bert_out)
        x = self.sigmoid(x)

        return x
```

```python
# Load the pre-trained tokenizer and model
BERT_MODEL = "roberta-base"
tokenizer = AutoTokenizer.from_pretrained(BERT_MODEL)
base_model = AutoModel.from_pretrained(BERT_MODEL)
```

```python
# Initialize the Transformer model and load weights
model = transformer(base_model)

# Save the entire model
torch.save(model, 'best_model.pt')

# Load the model's state dictionary with explicit map_location
loaded_model = torch.load('best_model.pt', map_location=lambda storage, loc: storage.cuda() if torch.cuda.is_available() else storage.cpu())

# Load state dict into the model
model.load_state_dict(loaded_model.state_dict())

# Move the model to the correct device
model.to(device)
```

```python
# Define criterion and optimizer
criterion = nn.BCELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=1e-5)
```

```python
# Instantiate the test dataset and dataloader
test_dataset = Dataset(test_data)
test_dataloader = DataLoader(test_dataset, batch_size=32, num_workers=4, collate_fn=collate_fn, pin_memory=True, worker_init_fn=seed_worker, generator=g)
```

**TESTING PHASE**

```python
from tqdm import tqdm
import torch
from sklearn.metrics import accuracy_score, roc_auc_score
import numpy as np

# Initialize metrics storage
total_loss_test = 0
all_labels = []
all_probs = []  # Storing probabilities instead of predictions
all_preds = []  # Storing binary predictions for accuracy calculation

# Wrap the loop over test_dataloader with tqdm for progress
for test_input, test_label in tqdm(test_dataloader, desc="Testing Progress", leave=False):
    attention_mask = test_input['attention_mask'].to(device)
    input_ids = test_input['input_ids'].squeeze(1).to(device)
    test_label = test_label.to(device)

    with torch.no_grad():
        output = model(input_ids, attention_mask)
        loss = criterion(output, test_label.float().unsqueeze(1))
        total_loss_test += loss.item()

        probs = torch.sigmoid(output)  # Use sigmoid to convert logits to probabilities
        preds = (probs >= 0.5).int()  # Convert probabilities to binary predictions
        all_labels.extend(test_label.cpu().numpy())
        all_probs.extend(probs.cpu().numpy())
        all_preds.extend(preds.cpu().numpy())

# Calculate the overall test loss
test_loss = total_loss_test / len(test_dataloader)

# Convert lists to numpy arrays for computation
all_labels = np.array(all_labels)
all_probs = np.array(all_probs)
all_preds = np.array(all_preds)

# Calculate AUC ROC using probabilities
if len(np.unique(all_labels)) == 2:
    auc_roc = roc_auc_score(all_labels, all_probs)
    test_accuracy = accuracy_score(all_labels, all_preds)
    print(f"Test Loss: {test_loss}")
    print(f"Test AUCROC: {auc_roc:.4f}")
    print(f"Test Accuracy: {test_accuracy:.4f}")
else:
    print("Warning: Only one class present in y_true. ROC AUC score is not defined in this case.")
    print(f"Test Loss: {test_loss}")
    print("Accuracy cannot be computed with a single class.")
```
